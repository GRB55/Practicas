```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
plt.style.use("ggplot")
```

```python
anual.dropna(how="all", inplace=True)
anual.reset_index(drop=True)
```

```python
anual.sort_values(by="Order Date", inplace=True)
anual.head()
```

```python
anual.columns
```

```python
anual.rename(columns={"Order ID": "Order_ID",
                     "Quantity Ordered": "Quantity_Ordered",
                     "Price Each": "Price_Each",
                     "Order Date": "Order_Date",
                     "Purchase Address": "Purchase_Address"}, inplace=True)
```

```python
anual.info()
```

```python
anual["Order_ID"] = pd.to_numeric(anual["Order_ID"])
anual["Quantity_Ordered"] = pd.to_numeric(anual["Quantity_Ordered"])
```

```python
anual.query("Order_ID == 'Order ID'")
```

```python
anual.replace(["Order ID", "Product", "Quantity Ordered", "Price Each", "Order Date", "Purchase Address"],
                 np.nan, inplace=True)
```

```python
anual.dropna(how="all", inplace=True)
```

```python
anual.reset_index(drop=True, inplace=True)
```

```python
anual.tail()
```

```python
anual["Price_Each"] = pd.to_numeric(anual["Price_Each"])
```

```python
anual["Order_Date"] = pd.to_datetime(anual["Order_Date"], format = "%m/%d/%y %H:%M")
```

```python
anual.info()
```

```python
anual["Total_Spent"] = anual["Quantity_Ordered"] * anual["Price_Each"]
```

```python
anual.head()
```

```python
anual.set_index("Order_Date", inplace=True)
```

```python
ventas = anual["Total_Spent"].resample("ME").sum()
ventas
```

```python
ventas.max()
```

```python
plt.figure(figsize=(10,6))
ventas.plot(kind="line",
            title="Ventas mensuales 2019",
            xlabel="",
            ylabel="Cantidad en millones de USD",
            marker="o",
            color="green",
            linewidth=2,
           )
plt.show()
```

```python
plt.figure(figsize=(10,6))
ventas.plot(kind="barh",
           title="Ventas mensuales 2019",
           ylabel="",
           xlabel="Cantidad en millones de USD",
            edgecolor="black",
           )

plt.show()
```

```python
anual[["Street", "City", "Postal_Code"]] = anual["Purchase_Address"].str.split(",", expand=True)
```

```python
anual.head()
```

```python
city_sales = anual.groupby("City")["Total_Spent"].agg(["sum"])
```

```python
city_sales.sort_values(by="sum", inplace=True)
```

```python
city_sales.plot(kind="barh",
               title="Ventas por ciudad",
               xlabel="Cantidad en millones USD",
               ylabel="",
               color="green",
               edgecolor="black",
               legend=False)
plt.show()
```

```python
city_sales2 = anual.groupby("City")["Total_Spent"].agg(["count"])
```

```python
city_sales2.sort_values(by="count", inplace=True)
```

```python
city_sales2.plot(kind="barh",
                title="Cantidad de productos vendidos por ciudad",
                xlabel="Cantidad",
                ylabel="",
                legend=False)

plt.show()
```

```python
juntos = anual.loc[anual["Order_ID"].duplicated(keep=False)]
juntos
```

```python
juntos["Grouped"] = juntos.groupby("Order_ID")["Product"].transform(lambda x: ",".join(x))
```

```python
juntos.head()
```

```python
juntos = juntos[["Order_ID","Grouped"]].drop_duplicates()
```

```python
juntos.head()
```

```python
from itertools import combinations
from collections import Counter

items = Counter()
for row in juntos["Grouped"]:
    rowlst = row.split(",")
    items.update(Counter(combinations(rowlst, 2)))
items
```

```python
prod = anual.groupby("Product")["Quantity_Ordered"].count()
prod.head()
```

```python
prices = anual.groupby("Product")["Price_Each"].mean()
prices
```

```python
prod.plot(kind="bar",
          ylabel="Cantidad",
          color="purple",
         legend=False,
)
prices.plot(kind="line",
            ylabel="Precio",
           secondary_y=True,
           rot=90)
plt.show()
```

```python
sns.scatterplot(x=prod, y=prices, c="brown")
plt.title("Relacion entre cantidad y precio")
plt.xlabel("Cantidad")
plt.ylabel("Precio")
plt.show()
```
