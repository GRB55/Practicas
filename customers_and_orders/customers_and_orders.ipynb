**Question 2:** In how many different states do the customers live in?

```python
display(customer_data["State"].nunique())

display(customer_data["State"].unique())
```

```python
states = customer_data["State"].unique()

states_clean = []

for state in states:
    clean_state = state.lower().strip()
    if clean_state not in states_clean:
        states_clean.append(clean_state)

display(states_clean)
display(len(states_clean))
```

### **ANSWER**
At first glance with the .nunique function we can observe 16 unique/different states but when we observe this values the are some inconsistencies:
- CA is different from Ca and (CA&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;) but they all are referring to California.
- After cleaning this mess we can see `there are 14 unique/different states where customers live in`.

### **RESUMEN**
A primera vista, cuando utilizamos la función .nunique podemos ver que hay 16 diferentes estados donde los clientes residen, pero cuando observamos los valores hay algunas inconsistencias:
- CA es diferente a Ca y (CA&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;) pero todos estos valores hacen referencia a California.
- Despúes de limpiar estos valores obtenemos que `hay 14 estados diferentes donde residen los clientes`.

**Question 5:** What is the most common last name?

```python
customer_data["LastName"].value_counts()
```

```python
last_names_count = {}

for last_name in customer_data["LastName"]:
    clean_last_name = last_name.strip().lower()
    if clean_last_name not in last_names_count:
        last_names_count[clean_last_name] = 1
    else:
        last_names_count[clean_last_name] += 1

last_name_df = pd.DataFrame(last_names_count.items(), columns=["LastName", "Count"])
last_name_df.sort_values(by="Count", ascending=False)
```

### **ANSWER**
We used the .value_counts function to get an idea of the possible answers but since we do not know the inconsistencies in the data we check each row individually and normalized. We achieve the same result for the answer:
`Smith is the most common last name`.

### **RESPUESTA**
Utilizamos la función .value_counts para tener una idea de las posibles respuestas pero como pueden existir inconsistencias en los datos también lo buscamos luego de normalizar las filas.
`Smith es el apellido más común/frecuente`.
